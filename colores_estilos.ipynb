```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,5,11)
```

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background')
```

```python
fig, ax=plt.subplots(figsize=(8,8))

ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'b-')
ax.plot(x,x+3,'y.-')
ax.plot(x,x+4,'go:') #we can use matlab style
```

## Pyplot

```python
fig, ax=plt.subplots(figsize=(8,8))

ax.plot(x,x+1,color='green', alpha=0.1,linewidth=8,linestyle='-',marker='o',markersize=15)
ax.plot(x,x+2,color='blue', linestyle='--',marker='x')
ax.plot(x,x+3,color='#ff8c00', linestyle='dashed', marker='v')
ax.plot(x,x+4,color='#a9a9a9', linestyle=':', marker='P') 
```
